```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.ticker import FuncFormatter
import time
```

```python
%matplotlib notebook
df=pd.read_csv("WHO-COVID-19-global-data.csv")
country_list=list(enumerate(df["Country"].unique()))
for i in country_list:
    print(i)
time.sleep(5)
country=input("Please write the number or the name of the country you want to select: ")
if country.isdigit()==True:
    if int(country)>236 or int(country)<0:
        print("The number you selected is not in range!")
        country=input("Please write the number or the name of the country you want to select: ")
    else:
        for j in country_list:
            a,b=j
            if str(a)==country:
                country=b
else:
    if country not in df["Country"].unique():
        print("You wrote the country name wrong! Please write country name again (case sensitive) or try using country number!")
        country=input("Please write the number or the name of the country you want to select: ")
df["Date_reported"]= pd.to_datetime(df["Date_reported"])
df=df.set_index("Date_reported")
df=df[df["Country"]==country]
df=df.drop(["WHO_region","Country","Country_code"],axis=1)

fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(2, 2,sharex=True)

axes=[ax1,ax2,ax3,ax4]
fig.set_figheight(7)
fig.set_figwidth(15)
fig.autofmt_xdate(rotation=90)
def human_format(num,pos):
    magnitude = 0
    while abs(num) >= 1000:
        magnitude += 1
        num /= 1000.0
    return '%.0f%s' % (num, ['', 'K', 'M'][magnitude])
numformat=FuncFormatter(human_format)

for i in range(4):
    axes[i].grid()
    axes[i].xaxis.set_major_formatter(mdates.DateFormatter('%m-%y'))
    axes[i].tick_params(bottom = False,left=False)
    axes[i].plot(df[df.columns[i]])
    axes[i].set_title(df.columns[i].replace("_"," ").title(),fontsize=10)
    axes[i].ticklabel_format(style='plain', axis='y')
    if i%2!=0:
        axes[i].yaxis.set_label_position("right")
        axes[i].yaxis.tick_right()
        axes[i].tick_params(right = False)
    axes[i].yaxis.set_major_formatter(numformat)

plt.suptitle(f'{country} Covid-19 Cases and Deaths',fontsize=16)
```
